## Univariate Analysis on Petal Length

```python
# PDF check for dataset overlap
sns.FacetGrid(df, hue='Species',height=6)\
    .map(sns.distplot,"Petal.Length")\
    .add_legend()
```

```python
# PDF check for dataset overlap
sns.FacetGrid(df, hue='Species',height=6)\
    .map(sns.distplot,"Petal.Width")\
    .add_legend()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
sns.boxplot(x='Species', y = 'Petal.Length', data=df)
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
sns.boxplot(x='Species', y = 'Petal.Width', data=df)
plt.show()
```
